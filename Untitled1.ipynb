```python
## basic optimize
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
f = lambda x: x[0]**2-2*x[0]*x[1]+4*x[1]**2-4*x[0]-4*x[1]+24
bnds =((0,4),(0,2))
res = minimize(f,(1,0), bounds=bnds)
print(res.x)
```

```python
# data that to fix for certain functional form FITTING CURVE
from scipy.optimize import curve_fit
x = np.linspace(0,5,30)
y = x**2 -4
plt.plot(x,y, 'o', label='parabolic curve')
plt.legend()
plt.show()   ## i want fit the curve
def f(x, a, b):
    return a*x**2+b
popt,pcov = curve_fit(f,x,y,p0=(1,1))
print(popt)
print(pcov)
```

```python
## special , differentiation , integration
from scipy.integrate import quad
import numpy as np
import matplotlib.pyplot as plt
integrand = lambda x: np.e**x+np.sin(x)**2

integral, intergral_error =quad(integrand, -1, 1)
print("definite integral is:", integral)
print(intergral_error)


## generating values for plot
x = np.linspace(-1,1,20)
y = integrand(x)
plt.plot(x,y,)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.fill_between(x,y, color='g')
plt.show()
```

```python
## newton law of colling
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from scipy.interpolate import interpld
time = np.array([0,1.0434,2.0868,3.1312,4.1756,5.2200,6.2644,7.3088,8.3478,9.3913,10.4347,11.4782,12.5217,13.5652,14.6086])
temp = np.array([283.2322,284.6945,286.2259,287.8306,289.6440,291.6187,293.7939,296.1148,298.4359,300.5430,302.1566,303.0363,303.0363,302.1566,300.5430])
print(temp.size)
print(time.size)
plt.scatter(time,temp, color = 'g')
plt.title("Newton law of cooling")
plt.xlabel("time[hour]")
plt.ylabel("temp[k]")
plt.show
temperature_surrounding = interpld(time,temp, kind = 'cubic')
def dTdt(T,t):
   return -0.5*(temp - temperature_surrounding(t))
times = np.linespace(0,14,100)
temp0 = 300
solve = odeint(dTdt, temp0, times)
```

```python
from scipy.interpolate import interp1d
temperature_surrounding = interp1d(time,temp, kind = 'cubic')
def dTdt(T,t):
   return -0.5*(T - temperature_surrounding(t))
times = np.linspace(0,14,100)
temp0 = 284.6945
solve = odeint(dTdt, temp0, times)
plt.plot(times, solve, c='g', label = 'temperature of the body')
plt.scatter(time, temp, color ='hotpink', label = 'temperature of surrounding')
plt.legend()
plt.title("Newton law of cooling")
plt.xlabel("time[hour]")
plt.ylabel("temp[k]")
plt.show()
```
